```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')
```

```python
df
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Create new columns
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
# Profit Margin
df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
# Which products generate the most revenue?
top_products= df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)
```

```python
top_products
```

```python
# How do sales change over time?
sales_over_time = df.groupby('Order Date')['Sales'].sum()
```

```python
sales_over_time
```

```python
#Which categories are most profitable?
profitable_categories = df.groupby('Category')['Profit'].sum()
```

```python
profitable_categories
```

```python
#Which region performs best?
best_region = df.groupby('Region')['Profit'].sum()
```

```python
best_region
```

```python
#Profit Margin by Category
profit_margin_cat = df.groupby('Category')['Profit Margin'].mean()
```

```python
profit_margin_cat
```
